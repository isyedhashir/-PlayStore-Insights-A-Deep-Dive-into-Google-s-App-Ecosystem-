```python
reviews_max = dataset.groupby("App")["Reviews"].max()
```

```python
reviews_max["Instagram"]
```

```python
# Create an empty  list to store clean data
dataset_clean = []


already_added = []


for index, row in dataset.iterrows():
    name = row["App"]
    n_reviews = row["Reviews"]
    
    if (reviews_max[name] == n_reviews ) and (name not in already_added):
        dataset_clean.append(row)
        already_added.append(name)
```

```python
dataset_clean = pd.DataFrame(dataset_clean)
```

```python
dataset_clean.shape
```

```python
def is_english(app_name):
    lst = []
    for i in app_name:
        if ord(i)  > 127:
            lst.append(False)
        else:
            lst.append(True)
    
    non_ascii = 0
    for j in lst:
        if j == False:
            non_ascii += 1
            
    if non_ascii > 3:
        return False
    else:
        return True
```
